```python
# 2667 단지번호붙이기 (version 1: DFS)

import sys

input = sys.stdin.readline

N = int(input().split()[0])
graph = [list(map(int, list(input().split()[0]))) for _ in range(N)]
visited = [[0]*N for _ in range(N)]
rst = []

def Dfs(R):
  queue = []
  queue.append(R)
  cnt = 0
  visited[R[1]][R[0]] = 1

  while queue:
    x,y = queue.pop()
    cnt += 1

    for dx,dy in zip([-1,0,0,1],[0,-1,1,0]):
      nx,ny = x+dx,y+dy

      if (0 <= nx <= N-1) and (0 <= ny <= N-1):
        if (visited[ny][nx] == 0) and (graph[ny][nx] == 1):
          queue.append([nx,ny])
          visited[ny][nx] = 1

  return cnt

for i in range(N):
  for j in range(N):
    if graph[j][i] == 1 and visited[j][i] == 0:
      rst.append(Dfs([i,j]))

rst.sort()
print(len(rst))
for i in range(len(rst)):
  print(rst[i])
```

```python
# 2667 단지번호붙이기 (version 2: BFS)

from collections import deque
import sys

input = sys.stdin.readline

N = int(input().split()[0])
graph = [list(map(int, list(input().split()[0]))) for _ in range(N)]
visited = [[0]*N for _ in range(N)]
rst = []

def Dfs(R):
  queue = deque()
  queue.append(R)
  cnt = 1
  visited[R[1]][R[0]] = 1

  while queue:
    x,y = queue.popleft()

    for dx,dy in zip([-1,0,0,1],[0,-1,1,0]):
      nx,ny = x+dx,y+dy

      if (0 <= nx <= N-1) and (0 <= ny <= N-1):
        if (visited[ny][nx] == 0) and (graph[ny][nx] == 1):
          queue.append([nx,ny])
          cnt += 1
          visited[ny][nx] = 1

  return cnt

for i in range(N):
  for j in range(N):
    if graph[j][i] == 1 and visited[j][i] == 0:
      rst.append(Dfs([i,j]))

rst.sort()
print(len(rst))
for i in range(len(rst)):
  print(rst[i])
```

```python
# 1068 트리 (트리는 부모 노드와 자식 노드로 이루어져서 모두 연결되어 있으며, 다른 부모 노드가 공통된 자식 노드를 가지는 경우가 없음)

import sys

input = sys.stdin.readline

N = int(input().split()[0])
graph = [[] for _ in range(N)]
structure = list(map(int, input().split()))

for i in range(len(structure)):

  if structure[i] == -1:
    continue

  graph[structure[i]].append(i)

rm = int(input())

R = structure.index(-1)

queue = []
queue.append(R)
cnt = 0

while queue:
  vert = queue.pop()

  if rm == R:
    break

  if len(graph[vert]) == 0:
    cnt += 1
    continue

  if graph[vert] == [rm]:
    cnt += 1
    continue

  for v in graph[vert]:

    if v == rm:
      continue

    else:
      queue.append(v)

print(cnt)
```

```python
# 1260 DFS와 BFS

from collections import deque
import sys

input = sys.stdin.readline

N, M, V = map(int, input().split())
graph = [[] for _ in range(N+1)]

for _ in range(M):
  vert1, vert2 = map(int, input().split())
  graph[vert1].append(vert2)
  graph[vert2].append(vert1)

def Dfs(start):

  queue = []
  queue.append(start)
  rst_Dfs = []
  visited = [0]*(N+1)

  while queue:

    vert = queue.pop()

    if visited[vert] == 1:
      continue

    visited[vert] = 1
    rst_Dfs.append(vert)
    graph[vert].sort(reverse=True)

    for v in graph[vert]:
      if visited[v] == 0:
        queue.append(v)

  return rst_Dfs

def Bfs(start):

  queue = deque()
  queue.append(start)
  rst_Bfs = [start]
  visited = [0]*(N+1)
  visited[start] = 1

  while queue:

    vert = queue.popleft()
    graph[vert].sort()

    for v in graph[vert]:
      if visited[v] == 0:
        queue.append(v)
        rst_Bfs.append(v)
        visited[v] = 1

  return rst_Bfs

rst1, rst2 = Dfs(V), Bfs(V)
print(*rst1)
print(*rst2)
```

```python
# 10026 적록색약 ([R],[G],[B]가 모두 다르게 인식되는 경우와, [R,G],[B]로 인식해야 하는 경우를 나눠서 각 색을 탐색하고 나머지 색은 벽으로 인식하도록 코드를 구성)(A in color를 이용해서 했음)

import sys

input = sys.stdin.readline

N = int(input().split()[0])
graph = [list(input().split()[0]) for _ in range(N)]

def Dfs(start,color):

  queue = []
  queue.append(start)

  while queue:
    x,y = queue.pop()

    if visited[x][y] != 0:
      continue

    visited[x][y] = 1

    for dx,dy in zip([-1,0,0,1],[0,-1,1,0]):
      nx,ny = x+dx,y+dy

      if (0 <= nx <= N-1) and (0 <= ny <= N-1):
        if (visited[nx][ny] == 0) and (graph[nx][ny] in color):
          queue.append([nx,ny])

  return 1

rst1 = []

for c in [['R'],['G'],['B']]:
  visited = [[0]*N for _ in range(N)]
  num = 0

  for i in range(N):
    for j in range(N):
      if visited[i][j] == 0 and graph[i][j] in c:
        num += Dfs([i,j],c)

  rst1.append(num)

rst2 = []

for c in [['R','G'],['B']]:
  visited = [[0]*N for _ in range(N)]
  num = 0

  for i in range(N):
    for j in range(N):
      if visited[i][j] == 0 and graph[i][j] in c:
        num += Dfs([i,j],c)
  rst2.append(num)

print(sum(rst1), sum(rst2))
```
